```python

Домашнее задание "Линейная алгебра. Векторы".¶
Даны векторы x = (1,1), y = (2, 0) и z = (0,2)

Надо:

Изобразить векторы на экране
Изобразить точку x + y + z
Найти угол между векторами x и y
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1., 1.])
y = np.array([2., 0.])
z = np.array([0., 2.])

plt.figure(figsize=(12, 7))
plt.plot(x[0], x[1], 'o', c = 'b')
plt.plot(y[0], y[1], 'o', c = 'r')
plt.plot(z[0], z[1], 'o', c = 'g')
plt.arrow(0, 0, 1, 1,
         length_includes_head = True,
         width = 0.01, color = 'b', label = 'x' )
plt.arrow(0, 0, 2, 0,
         length_includes_head = True,
         width = 0.01, color = 'r', label = 'y')
plt.arrow(0, 0, 0, 2,
         length_includes_head = True,
         width = 0.01, color = 'g', label = 'z')

plt.legend()
plt.grid(True)
```

```python
from numpy import linalg as LA
u = np.degrees(np.arccos(np.dot(x, y)/(LA.norm(x, ord = 2)*LA.norm(y, ord = 2))))
print(f' Угол равен: {u:.2f} градусов')
```
