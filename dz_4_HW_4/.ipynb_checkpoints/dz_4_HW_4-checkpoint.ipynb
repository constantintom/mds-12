### Уровень 0:


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
from sympy import *

x = symbols('x')
exp = x**5 + 4*sin(2*x) + cos(3*x + 3)
print("До диффиренцирования: {}".format(exp))
dif1 = diff(exp, x)
print("Первая производная : {}".format(dif1))
print(dif1.subs(x, 1).evalf())
dif2 = diff(dif1, x)
print()
print("Вторая производная : {}".format(dif2))
print(dif2.subs(x, 1).evalf())
print()
def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x + 3)

print(f' 1-я: {derivative(f, 1.0, dx=1e-6)}')
print(f' 2-я: {derivative(f, 1.0, dx=1e-6, n = 2)}')
```

### Уровень 1:


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
a = np.linspace(-5, 5) # Создаем равномерно распределённый массив
from scipy.signal import find_peaks
def f(x):
    return (np.sin(2*x + 1))**5
df = [f(x) for x in a]
df1 = [derivative(f, x, dx=1e-6, n = 1) for x in a]
df2 = [derivative(f, x, dx=1e-6, n = 2) for x in a]
fig, axes = plt.subplots(figsize = (14, 8))
plt.title('График функции и её производных')
axes.set_xlabel('Аргументы')
axes.set_ylabel('Ординаты')
axes.plot(a, df, 'r', lw = 3, label = 'Функция')
axes.plot(a, df1, 'b-.', lw = 2, label = '1-я производная')
axes.plot(a, df2, 'g--', lw = 2, label = '2-я производная')
axes.legend(fontsize=12)
plt.show
```
