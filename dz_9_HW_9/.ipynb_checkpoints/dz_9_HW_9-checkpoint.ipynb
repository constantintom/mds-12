```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
rv = sts.gumbel_l
sample = rv.rvs(loc = 0., scale = 1.0, size = 1000)
```

```python
x = np.linspace(-8,2,100)
pdf = rv.pdf(x)
fig, axes = plt.subplots(figsize = (13, 6))
axes.hist(sample, density=True, bins=30, label='sample histogram')
axes.plot(x, pdf,  'r', lw = 8, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
# Параметры запуска: размеры выборок и количество повторов генерации
n1 = 2
n2 = 10
n3 = 50
samples_count = 1000
# Массивы для сохранения средних по выборкам
a1 = np.array([])
a2 = np.array([])
a3 = np.array([])

for number in range(samples_count):
    # На каждом цикле генерируем выборки разных размеров
    sample1 = rv.rvs(size=n1)
    sample2 = rv.rvs(size=n2)
    sample3 = rv.rvs(size=n3)   
    # Считаем среднее по каждой выборке и добавляем в массив средних
    a1 = np.append(a1, sample1.mean())
    a2 = np.append(a2, sample2.mean())
    a3 = np.append(a3, sample3.mean())    
```

```python
# Строим графики
# Подписи для графиков
fig, axes = plt.subplots(1, figsize=(11,6))
label_1 = 'sample of ' + str(n1) + ' histogram'
axes.hist(a1, density=True, label=label_1, bins = 30)
plt.legend()
plt.ylabel('samples means')
plt.xlabel('$x$')
```

```python
fig, axes = plt.subplots(1, figsize=(11, 6))
label_2 = 'sample of ' + str(n2) + ' histogram'
axes.hist(a2, density=True, label=label_2, bins = 30, color = 'g')
plt.legend()
plt.ylabel('samples means')
plt.xlabel('$x$')
```

```python
fig, axes = plt.subplots(1, figsize=(11,6))
label_3 = 'sample of ' + str(n3) + ' histogram'
axes.hist(a3, density=True, label=label_3, bins = 30, color = 'r')
plt.legend()
plt.ylabel('samples means')
plt.xlabel('$x$')
```
